### Data extraction

```python
import pandas as pd

data = pd.read_excel(r"C:\Users\AgJo413\Documents\GitHub\Lab_std\labstds\Data\Boston Housing Dataset 1978.xlsx")

data.head()
```

```python
data.describe()
```

```python
data_filtered = data.dropna()

data_filtered.describe()
```

```python
import numpy as np 

x = data_filtered.iloc[:, 0:-1].values
X = np.hstack((x, np.ones([len(x[:, 0]), 1])))
print(len(X[0, :]))

y = data_filtered.iloc[:, -1]
```

## Model

```python
from sklearn.linear_model import LinearRegression 

model = LinearRegression()

model.fit(X, y)
betas = [model.intercept_]

for i in model.coef_:
    betas.append(i)
    
print("Betas: {}".format(betas))
```

```python
n = len(X)
p = len(betas)

ypred = model.predict(X)

rss = np.sum((y - ypred) ** 2)
var = rss / (n - p)
std = np.sum((x - np.average(x) ** 2))


```

```python
import statsmodels.api as sm 
ols = sm.OLS(y, X)
results = ols.fit()
results.summary()
```

```python
data_filtered = data_filtered.drop(data_filtered.columns[[0, 2, 6]], axis=1)
data_filtered.head()
```

```python
for i in data_filtered.columns[:-1]:
    data_filtered[i] = data_filtered[i] - data_filtered[i].min()
    data_filtered[i] = data_filtered[i] / data_filtered[i].max()
data_filtered.describe()  
```

```python
x = data_filtered.iloc[:, 0:-1].values
X = np.hstack((x, np.ones([-1, 1])))
print(len(X[0, :]))

y = data_filtered.iloc[:, -1]

ols = sm.OLS(y, X)
results = ols.fit()
results.summary()
```
