```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
import numpy as np 

data = pd.read_csv(r"C:\Users\AgJo413\Documents\GitHub\Lab_std\labstds\Data\imdb_top1000_lae.csv")


data.head()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
data.describe()
```

```python
dataf = data.drop(data.columns[[0, 1]], axis=1)
dataf.head()
```

```python
dataf['Gross'] = dataf['Gross'].str.replace(",", "")
dataf["Gross"] = pd.to_numeric(dataf.loc[:, "Gross"], errors='coerce')
```

```python
dataf["Runtime"] = dataf["Runtime"].str.replace(" min", "")
dataf["Runtime"] = pd.to_numeric(dataf.loc[:, "Runtime"], errors='coerce')
```

```python
dataf.head(1)
```

```python
categoric = ["Certificate", "Director"]
encoded_features = pd.get_dummies(dataf[categoric],
                                  columns=categoric,
                                  drop_first=False)
```

```python
encoded_features
```

```python
category = dataf.iloc[:, 7:].astype(bool)
category = category.rename(columns=lambda x: "Category_{}".format(x))
category
```

```python


dataf["Released_Year"] = pd.to_numeric(dataf.loc[:, "Released_Year"], errors='coerce') / 1000
dataf
```

```python
numeric = ['Released_Year', 'Runtime', 'Meta_score','Gross']
```

```python


dataf = pd.DataFrame(dataf[numeric], 
                                columns=numeric).reset_index()

encoded_features = encoded_features.reset_index()

dataf = dataf.merge(encoded_features, on='index')

dataf = dataf.drop('index', axis=1)


dataf
```

```python
x = dataf
y = data["IMDB_Rating"]
x.describe()
```

```python
def modeling(trainx, trainy, testx, testy):
    model = LinearRegression()
    model.fit(trainx, trainy)
    ypred = model.predict(testx)
    return ypred

def MSE(ypred, testy):
    return mean_squared_error(testy, ypred)

data_shuffled = dataf.sample(frac=1)
```

```python
mses = []
for i in range(0, len(data_shuffled) - 4, 10):
    testx = x.iloc[i:i+10, :]
    testy = y.to_frame().iloc[i:i+10, :]
    trainx = x.drop(testx.index)
    trainy = y.to_frame().drop(testx.index)
    ypred = modeling(trainx, trainy, testx, testy)
    mses.append(MSE(ypred, testy))

print("MSE has an average of: {}".format(np.mean(mses)))
print("MSE has a standard deviation of: {}".format(np.std(mses)))
    
```
